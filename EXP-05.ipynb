```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def generate_data(n=100):
    np.random.seed(42)
    X = np.arange(n).reshape(-1, 1)
    y = 5 * X.ravel() + np.random.normal(scale=10, size=n)
    return X, y


X, y = generate_data()


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred_full = model.predict(X)

mse = mean_squared_error(y_test, model.predict(X_test))
print(f"Mean Squared Error: {mse:.2f}")

plt.figure(figsize=(10, 5))
plt.scatter(X, y, label="Actual Data", color="blue", alpha=0.5)  # Original data points
plt.plot(X, y_pred_full, label="Linear Trend Line", color="red", linewidth=2.5)  # Full prediction line
plt.xlabel("Time")
plt.ylabel("Value")

plt.legend()
plt.title("Linear Regression for Time Series Forecasting")
plt.show()
```
